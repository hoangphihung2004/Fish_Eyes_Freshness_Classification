```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
def process_directory(image_paths):
    all_paths = []
    for root, dirs, files in os.walk(image_paths):
        tmp = root.split("\\")
        if files:
            fish_name = tmp[-1]
            label = tmp[-1].split(" - ")[1].strip()
            for file in files:
                if file.endswith((".jpg", ".png", ".JPG", ".PNG")):
                    all_paths.append({"Fish Name": fish_name, "Label": label, "Path": file})
    return pd.DataFrame(all_paths)
```

```python
df = process_directory(r".\Fish_Eyes_Freshness_Classification\data\xzyx7pbr3w-1")
df["Fish Name + Label"] = df["Fish Name"] + " " + df["Label"]

df
```

```python
x = df.drop(columns=["Fish Name + Label"])
y = df["Fish Name + Label"]
print(x.shape)
print(y.shape)
```

```python
# 64:16:20
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
print(sum(y_train.value_counts().values) / 4390)
print(sum(y_val.value_counts().values) / 4390)
print(sum(y_test.value_counts().values) / 4390)
```

```python
train = pd.concat([x_train, y_train], axis=1, ignore_index=True)
val = pd.concat([x_val, y_val], axis=1, ignore_index=True)
test = pd.concat([x_test, y_test], axis=1, ignore_index=True)
train.columns = df.columns
val.columns = df.columns
test.columns = df.columns
train = train.drop(columns=["Fish Name + Label"])
val = val.drop(columns=["Fish Name + Label"])
test = test.drop(columns=["Fish Name + Label"])
train["Type"] = "Train"
val["Type"] = "Validation"
test["Type"] = "Test"
print(train.shape)
print(val.shape)
print(test.shape)
```

```python
new_df = pd.concat([train, val, test], ignore_index=True)
new_df
```

```python
new_df.to_csv(r"D:\Fish_Eyes_Freshness_Classification\Data\meta_data.csv", index=False)
```
